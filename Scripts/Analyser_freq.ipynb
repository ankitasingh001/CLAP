```python
import codecs
import re
from collections import defaultdict
from glob import glob
import matplotlib.pyplot as plt
import os
from pathlib import Path
import csv
```

```python
# Check if the name is a language

def check_lang(lang):
    if lang =='Hindi' or lang=='Bengali' or lang=='Odia' or lang=='Tamil' or lang =='Telugu' or lang == 'Marathi' or lang == 'Malayalam' or lang == 'Punjabi' or lang=='Gujrati' or lang=='Kannada':
        return True
    else:
        return False
```

```python
# Seperator for various languages

def seperator(language):
    if language =='Hindi' or language =='Bengali' or language=='Odia' or language =='Punjabi':
        return u"।"
    else: 
        return "."
```

```python
# generate csv file from observed data

def write_list_to_file(guest_list, filename):
    with open(filename,'a') as outfile:
        wr = csv.writer(outfile, dialect='excel')
        wr.writerow(guest_list)
```

```python
# Necessary methods 

def clean_text(text):
    '''not working'''
    text = re.sub(r'(\d+)', r'', text)
    text = text.replace(u',', '')
    text = text.replace(u'"', '')
    text = text.replace(u'(', '')
    text = text.replace(u')', '')
    text = text.replace(u'"', '')
    text = text.replace(u':', '')
    text = text.replace(u"'", '')
    text = text.replace(u"‘‘", '')
    text = text.replace(u"’’", '')
    text = text.replace(u"''", '')
    text = text.replace(u".", '')
    text = text.replace(u"।",' ')
    return text

def print_sentences(sentence):
    for i in sentences: 
        print (i.encode('utf-8'))

def print_token(tokens):
    for i in tokens: 
        print (i.encode('utf-8'))

def print_freq(freq):
    for i in sorted(freq.key()):
        print (i,',',freq[i])
```

```python
# Defining list of languages

language_list = ['Hindi','Bengali','Tamil','Telugu','Marathi','Malayalam','Gujrati','Punjabi','Kannada','Odia']
```

```python
# Initialising variables and paths

hist ={}
for lang in language_list:
    hist[lang] = defaultdict(int)

sentences =[]
orig_path = "/home/ankita/CLAP/Data_copy"
new_path ="/home/ankita/CLAP/Data_2/"
new_data_path = "/home/ankita/CLAP/data_1"
csv_path = "/home/ankita/CLAP/Data/after_Filter_summary.csv"
```

```python
# Method to calculate frequency of every word in the file for each language

def frequency_calc(path,language,seperator):
    i=0
    length=0
    for file in Path(path).glob('**/*.txt'):
        sentences =[]
        tokens =[]
        with codecs.open(file, encoding='utf-8') as f:
            txt = f.read()
            sentences = sentences + txt.split(seperator)
            length = length + len(sentences)
            clntxt = clean_text(txt)
            tokens = clntxt.split(' ')
            for token in tokens :
                hist[language][token] += 1
            i=i+1
```

```python
# Method to Count and write the filtered sentences 

def counter_afterfilter(read_path,write_path,language):
     
    flag =0
    flength1=0
    flength2=0
    flength3=0
    txt_file= open(read_path)
    f1 =open(write_path,"w+")
    for line in txt_file:
        sentences =[]
        filteredsent =""
        sentences = line.split(seperator(language))
        for sentence in sentences:
            tokens = sentence.split(' ')
            flag =0
            for token in tokens:
                if token in filteredtok1[language].keys():
                    flag = 1
                    break
            if flag == 0 and sentence != '' and sentence != '\n':
                flength1 = flength1 + 1
            flag =0
            for token in tokens:
                if token in filteredtok2[language].keys():
                    flag = 1
                    break
            if flag == 0 and sentence != '' and sentence != '\n' :
                flength2 = flength2 + 1
            flag =0
            for token in tokens:
                if token in filteredtok3[language].keys():
                    flag = 1
                    break
            if flag == 0 and sentence != ' ' and sentence != '\n' :
                filteredsent = sentence + seperator(language)
                flength3 = flength3 + 1
                f1.write(filteredsent)

    f1.close() 
    return flength1,flength2,flength3
```

```python
#Populating dictionary for all languages

Data_source = [f.path for f in os.scandir(orig_path) if f.is_dir()]

for data_path in Data_source:
    
    source =data_path.split('/')[-1]
    lang_for_source = [f.path for f in os.scandir(data_path) if f.is_dir()]
    
    for lang_path in lang_for_source:
        
        language =lang_path.split('/')[-1]        
        if(check_lang(language) is False):
            continue
        frequency_calc(lang_path,language,seperator(language))
```

```python
# Setting the filters for words in frequency counts 1 ,2 and 3

filteredtok1 = {}
filteredtok2 = {}
filteredtok3 = {}
for lang in language_list:
    filteredtok1[lang] = dict()
    filteredtok2[lang] = dict()
    filteredtok3[lang] = dict()
# Iterate over all the items in dictionary and filter items which has even keys
for (key, value) in hist.items():
   # Check if key is even then add pair to new dictionary
    if isinstance(value, dict):
        for inner_key, inner_value in value.items():
            print(inner_key,inner_value)
            if inner_value <=1 :
                filteredtok1[key][inner_key] = inner_value
            if inner_value<=2 :
                filteredtok2[key][inner_key] = inner_value
            if inner_value<=3 :
                filteredtok3[key][inner_key] = inner_value
```

```python
Data_source = [f.path for f in os.scandir(new_data_path) if f.is_dir()]

for data_path in Data_source:
    
    source =data_path.split('/')[-1]
    lang_for_source = [f.path for f in os.scandir(data_path) if f.is_dir()]
    
    for lang_path in lang_for_source:
        tot_len1 = 0
        tot_len2 = 0
        tot_len3 = 0
        language = lang_path.split('/')[-1]        
        if(check_lang(language) is False):
            continue
        for file in Path(lang_path).glob('**/*.txt'):
            # Create a path to write the files after filter
            
            file_name = str(file).split('/')[-1]
            filter_path = os.path.join(new_path, source, language)
            if not os.path.isdir(filter_path):
                os.makedirs(filter_path)

            len1,len2,len3 = counter_afterfilter(file,filter_path+'/'+file_name,language)
            tot_len1 += len1
            tot_len2 += len2
            tot_len3 += len3
        data_list = [source,language,tot_len1,tot_len2,tot_len3]
        write_list_to_file(data_list,csv_path)
```
